```
# Importação das Bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Leitura do Dataset Original

credit=pd.read_csv('dados/credit.csv')
#credit=credit.drop('clientid',axis=1)
credit=credit.set_index('clientid')

```

```
credit.columns.tolist()
#[clientid	income	age	loan	default]
```

```
credit.head(100)
```

```
credit.corr()
```

```
# Correlação entre as variáveis do Dateset
# (Pandas)
# DataFrame.corr(self, method='pearson', min_periods=1) → 'DataFrame
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html

types = credit.iloc[:, :12]
corr_types = pd.DataFrame.corr(types,method='pearson')
print(corr_types)
```

```
# Matriz de Correlação
# Numpy

corrs = credit.corr()
mask = np.zeros_like(corrs)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corrs, cmap='Spectral_r', mask=mask, square=True, vmin=-.4, vmax=.4)
plt.title('Matriz de Correlação')
```

```
# FacetGrid do Seaborn para criar uma scatterplot dos atributos age e loan.

sns.FacetGrid(credit, height=7).map(plt.scatter, "age", "loan")
```

```
# FacetGrid do Seaborn para criar uma scatterplot dos atributos income e loan.

sns.FacetGrid(credit, height=5).map(plt.scatter, "income", "loan") 
```

```
# Seaborn com os atributos age e income.

sns.jointplot(x="age", y="income", data=credit, height=5)
```
